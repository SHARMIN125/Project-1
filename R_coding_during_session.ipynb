```r
# Load necessary libraries
library(caret)
library(rpart)

# Check the working directory
getwd()

# Load the RDS model
model <- readRDS("height_weight_model.rds")

# new code

cat("Enter height (in CM): ")
user_height <- as.numeric(readline())


cat("Enter weight (in KG): ")
user_weight <- as.numeric(readline())


userInput <- data.frame(Height = user_height, Weight = user_weight)
userPrediction <- predict(model, userInput, type = "class")
cat("Prediction:\n")
cat("Height:", user_height, "CM, Weight:", user_weight, "KG -> Result:", as.character(userPrediction), "\n")




```
